# Техническое задание

## Проект: Прогнозирование оттока клиентов банка

---

# 1. Описание проекта

Разработать модель машинного обучения, которая будет прогнозировать, уйдёт клиент из банка или останется.

Модель должна решать задачу **бинарной классификации**:

- **0 / Stayed** — клиент останется в банке
- **1 / Exited** — клиент уйдёт из банка

Проект направлен на демонстрацию применения методов машинного обучения для решения практической банковской задачи.

# 2. Бизнес-цель проекта

Создать систему прогнозирования оттока клиентов, которая поможет банку заранее выявлять клиентов с высоким риском ухода.

Система может использоваться для:

- удержания клиентов;
- персональных предложений;
- анализа причин ухода;
- повышения качества банковского обслуживания;
- снижения финансовых потерь банка.

# 3. Основные задачи проекта

## Этап 1. Подготовка данных

- Найти и скачать реальный датасет **Bank Customer Churn Dataset**
- Изучить структуру данных
- Проверить типы признаков
- Проверить наличие пропущенных значений
- Проверить наличие дубликатов
- Проверить баланс классов
- Подготовить данные для дальнейшего анализа

## Этап 2. Исследовательский анализ данных (EDA)

Провести анализ:

- распределения целевой переменной `Exited`;
- возраста клиентов;
- кредитного рейтинга;
- баланса на счёте;
- количества используемых продуктов;
- активности клиента;
- влияния страны и пола клиента на отток.

Построить визуализации:

- Countplot для целевой переменной;
- Histplot для числовых признаков;
- Boxplot для анализа выбросов;
- Heatmap корреляции;
- Feature distribution по классам.

## Этап 3. Предобработка данных

Необходимо выполнить:

- удаление ненужных колонок, например `RowNumber`, `CustomerId`, `Surname`;
- обработку категориальных признаков;
- One-Hot Encoding для `Geography` и `Gender`;
- нормализацию числовых признаков;
- разделение данных на:
  - train
  - validation
  - test

## Этап 4. Построение baseline-моделей

Разработать и сравнить несколько базовых моделей:

- Logistic Regression
- Decision Tree
- Random Forest

Цель этапа — получить начальный результат и выбрать направление для улучшения модели.

## Этап 5. Построение улучшенной модели

Реализовать и сравнить более сильные модели:

- Gradient Boosting
- XGBoost
- CatBoost

Провести подбор параметров:

- `n_estimators`
- `max_depth`
- `learning_rate`
- `class_weight`
- `random_state`

## Этап 6. Fine-Tuning

Выполнить улучшение модели:

- подобрать гиперпараметры через GridSearchCV или RandomizedSearchCV;
- проверить модель на validation/test выборке;
- сравнить метрики разных моделей;
- выбрать лучшую модель;
- сохранить лучшую модель в файл.

## Этап 7. Оценка модели

Рассчитать метрики:

- Accuracy
- Precision
- Recall
- F1-score
- ROC-AUC

Построить:

- Confusion Matrix
- ROC Curve
- Feature Importance
- таблицу сравнения моделей

# 4. Функциональные требования

## Модель должна:

- принимать данные клиента на вход;
- анализировать характеристики клиента;
- возвращать вероятность ухода клиента из банка;
- возвращать итоговый класс:
  - **Stayed**
  - **Exited**

## Пользователь должен иметь возможность:

- ввести данные нового клиента;
- получить прогноз модели;
- увидеть вероятность оттока;
- получить понятное текстовое объяснение результата.

# 9. Критерии успешности

Минимальные целевые показатели:

- Accuracy ≥ 85%
- ROC-AUC ≥ 0.85
- F1-score ≥ 0.80

Дополнительно важно учитывать метрику **Recall**, так как банку важно не пропустить клиентов, которые действительно могут уйти.
